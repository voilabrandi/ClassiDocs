```python
#Evaluation results

#Import libraries
from pathlib import Path
import pandas as pd

# Define project directories
base_dir = Path.cwd().parent
data_dir = base_dir / "data"
output_dir = base_dir / "output"
output_dir.mkdir(parents=True, exist_ok=True)


# Load the main experiment log if it exists
log_path = output_dir / "ft_experiment_logs.csv"
log_df = pd.read_csv(log_path)



registry_path = output_dir / "ft_model_registry.csv"
model_registry_df = pd.read_csv(registry_path)




# Keep only the most important columns for reporting
summary_cols = ["config", "iter", "added_unique", "train_pool_size",
        "acc", "f1_macro", "f1_qual", "f1_quant", "model_path"
    ]
    
# Retain only columns that actually exist in the file
available_cols = [col for col in summary_cols if col in log_df.columns]
summary = log_df[available_cols].copy()

print("Recent experiment records:")
display(summary.tail(20))

    # Identify the best checkpoint for each configuration
if "config" in summary.columns and BEST_METRIC in summary.columns:
    per_cfg_best_idx = summary.groupby("config")[BEST_METRIC].idxmax()
    per_cfg_best = summary.loc[per_cfg_best_idx].sort_values(BEST_METRIC, ascending=False)

    print("\nBest checkpoint per config:")
    display_cols = [col for col in ["config", "iter", BEST_METRIC, "acc", "model_path"] if col in per_cfg_best.columns]
    display(per_cfg_best[display_cols])

    # Identify the overall best checkpoint across all configurations
    best_row = summary.loc[summary[BEST_METRIC].idxmax()]

    print("\nOverall best checkpoint")
    if "config" in best_row.index:
        print("config:", best_row["config"])
    if "iter" in best_row.index:
        print("iter:", int(best_row["iter"]))
    if BEST_METRIC in best_row.index:
        print(f"{BEST_METRIC}:", round(float(best_row[BEST_METRIC]), 4))
    if "model_path" in best_row.index:
        print("model_path:", best_row["model_path"])


 
```
